## **Taller de Complejidad Computacional**

### A) Suma de los n primeros números (iterativo)

Dado un número n, desarrolla una función que sume los n primeros números por medio de ciclos (iterativo).

**Solución:**

```python
# Importar librerías necesarias
import time
import pandas as pd
import matplotlib.pyplot as plt
```

### **Paso 1:**
Implementar el algoritmo iterativo.

Primero, implementamos una función que suma los n primeros números utilizando un ciclo for.

```python
def suma_iterativa(n):
    suma = 0
    for i in range(1, n + 1):
        suma += i
    return suma
```

### **Paso 2:** 
Medir el tiempo de ejecución
Para medir el tiempo de ejecución, usamos la librería time de Python. 
La función time.time() nos permitirá obtener el tiempo de inicio y fin, calculando el tiempo total de ejecución.

```python
def medir_tiempo(func, n):
    start_time = time.time()  # Comienza el temporizador
    func(n)  # Ejecuta la función
    end_time = time.time()  # Detiene el temporizador
    return end_time - start_time  # Calcula el tiempo de ejecución en segundos


```

### **Paso 3:** 
Ejecutar el programa con entradas de diferentes tamaños
Probamos el algoritmo con diferentes tamaños de entrada n **(n = 10, 100, 1000, 10000, 100000, 1000000, 10000000)** y guardamos los tiempos de ejecución para cada tamaño.

```python
valores_n = [10, 100, 1000, 10000, 100000, 1000000, 10000000]
tiempos_iterativo = []

# Ejecutar para cada tamaño de entrada
for n in valores_n:
    tiempo_iterativo = medir_tiempo(suma_iterativa, n)
    tiempos_iterativo.append(tiempo_iterativo)
    print(f"n={n} -> Iterativo: {tiempo_iterativo:.6f} segundos")
```

### **Paso 4:** 
Exportar los resultados a un archivo CSV

```python
df = pd.DataFrame({
    'n': valores_n,
    'Tiempo Iterativo': tiempos_iterativo
})

# Guardar el DataFrame en un archivo CSV
df.to_csv('resultados_complejidad.csv', index=False)

# Mostrar los primeros resultados
df.head()
```

### **Paso 5:** 

Graficar los resultados

```python
plt.figure(figsize=(10, 6))

plt.plot(valores_n, tiempos_iterativo, color='blue', label='Iterativo', marker='o', linestyle='-', markersize=6)

# Configuración del gráfico
plt.xscale('log')  # Escala logarítmica para el eje X
plt.yscale('log')  # Escala logarítmica para el eje Y
plt.xlabel('Tamaño de entrada (n)', fontsize=12)
plt.ylabel('Tiempo de ejecución (segundos)', fontsize=12)
plt.title('Comparación de tiempos de ejecución para la suma iterativa', fontsize=14)
plt.legend()

# Mostrar el gráfico
plt.show()
```

### B) Suma de los n primeros números (Gauss)

Dado un número n, desarrolla una función que sume los n primeros números por medio de la suma de Gauss.

**Solución:**

### **Paso 1:**
Implementar la función de Suma de Gauss

```python
def suma_gauss(n):
    return n * (n + 1) // 2  # Usamos la fórmula de Gauss
```

### **Paso 2:** 
Medir el tiempo de ejecución

```python

def medir_tiempo(func, n):
    start_time = time.time()  # Comienza el temporizador
    func(n)  # Ejecuta la función
    end_time = time.time()  # Detiene el temporizador
    return end_time - start_time  # Calcula el tiempo de ejecución en segundos
```

### **Paso 3:** 
Ejecutar el programa con diferentes tamaños de entrada

```python
tiempos_gauss = []

# Ejecutar para cada tamaño de entrada
for n in valores_n:
    tiempo_gauss = medir_tiempo(suma_gauss, n)
    tiempos_gauss.append(tiempo_gauss)
    print(f"n={n} -> Gauss: {tiempo_gauss:.6f} segundos")
```

### **Paso 4:** 
Exportar los resultados a un archivo CSV

```python

df_gauss = pd.DataFrame({
    'n': valores_n,
    'Tiempo Gauss': tiempos_gauss
})

# Guardar el DataFrame en un archivo CSV
df_gauss.to_csv('resultados_complejidad_gauss.csv', index=False)

# Mostrar los primeros resultados
df_gauss.head()
```

### **Paso 5:** 
Graficar los resultados con líneas

```python

# Paso 5: Graficar los resultados con líneas
plt.figure(figsize=(10, 6))

# Graficar la suma de Gauss
plt.plot(valores_n, tiempos_gauss, color='green', label='Gauss', marker='x', linestyle='-', markersize=6)

# Configuración del gráfico
plt.xscale('log')  # Escala logarítmica para el eje X
plt.xlabel('Tamaño de entrada (n)', fontsize=12)
plt.ylabel('Tiempo de ejecución (segundos)', fontsize=12)
plt.title('Tiempo de ejecución para la suma de Gauss', fontsize=14)
plt.legend()

# Mostrar el gráfico
plt.show()
```

# **COMPARACION DE TIEMPOS:**
Suma Iteractiva vs Gauss

```python


# Función para la suma iterativa
def suma_iterativa(n):
    suma = 0
    for i in range(1, n + 1):
        suma += i
    return suma

# Función para la suma de Gauss
def suma_gauss(n):
    return n * (n + 1) // 2  # Usamos la fórmula de Gauss


```

###  **Medir el tiempo de ejecución de una función**

```python

def medir_tiempo(func, n):
    start_time = time.time()  # Comienza el temporizador
    func(n)  # Ejecuta la función
    end_time = time.time()  # Detiene el temporizador
    return end_time - start_time  # Calcula el tiempo de ejecución en segundos
```

### **Tamaños de entrada n**

```python

valores_n = [10, 100, 1000, 10000, 100000, 1000000, 10000000]
```

### **Medir los tiempos de ejecución para cada n**

```python

tiempos_iterativo = []
tiempos_gauss = []  
for n in valores_n:
    tiempo_iterativo = medir_tiempo(suma_iterativa, n)
    tiempos_iterativo.append(tiempo_iterativo)
    
    tiempo_gauss = medir_tiempo(suma_gauss, n)
    tiempos_gauss.append(tiempo_gauss)
```

### **Grafico comparativo:**

```python

plt.figure(figsize=(10, 6))

# Graficar la suma iterativa
plt.plot(valores_n, tiempos_iterativo, color='blue', label='Iterativo', marker='o', linestyle='-', markersize=6)

# Graficar la suma de Gauss
plt.plot(valores_n, tiempos_gauss, color='green', label='Gauss', marker='x', linestyle='-', markersize=6)

# Configuración del gráfico
plt.xlabel('Tamaño de entrada (n)', fontsize=12)
plt.ylabel('Tiempo de ejecución (segundos)', fontsize=12)
plt.title('Comparación de tiempos de ejecución (Suma Iterativa vs Gauss)', fontsize=14)
plt.legend()

# Mostrar el gráfico
plt.show()
```

### **Análisis Empírico**

El análisis empírico se basa en la observación directa del comportamiento de los algoritmos a través de la medición de su **tiempo de ejecución**. Para este caso, se probaron dos métodos para sumar los primeros `n` números:

1. **Método Iterativo (Suma con ciclos)**: Un algoritmo que usa un bucle para sumar los primeros `n` números.
2. **Método de Gauss (Fórmula cerrada)**: Un algoritmo que usa la famosa fórmula de Gauss para sumar los primeros `n` números, \( S(n) = \frac{n(n+1)}{2} \), que tiene una complejidad constante.

#### **Procedimiento**

1. **Implementación**:
   - Para cada uno de los métodos, se implementaron funciones en Python que suman los primeros `n` números. Para medir el tiempo de ejecución, se utilizó la función `time()` de Python.
  
2. **Entradas de prueba**:
   - Se probaron diferentes valores de `n`: \( n = 10, 100, 1000, 10000, 100000, 1000000, 10000000 \).
   
3. **Medición de tiempo**:
   - Se midió el tiempo que cada algoritmo tardó para resolver el problema con cada valor de `n`.

4. **Análisis de los resultados**:
   - Los resultados se guardaron y graficaron. El gráfico final comparó los tiempos de ejecución de ambos métodos usando una escala lineal en ambos ejes.

#### **Resultados Empíricos**

1. **Método Iterativo**:
   - La **suma iterativa** tiene una complejidad **lineal** \( O(n) \), por lo que su tiempo de ejecución aumenta proporcionalmente con el tamaño de `n`.
   - Al graficar los tiempos de ejecución para diferentes tamaños de `n`, se observa un **crecimiento lineal**. En valores pequeños de `n`, el tiempo es bajo, pero conforme `n` aumenta (por ejemplo, \( n = 10^6 \) o \( n = 10^7 \)), el tiempo de ejecución también crece de manera más notoria.
   - En el gráfico de dispersión, el tiempo de ejecución de la suma iterativa muestra un aumento constante en relación con `n`.

2. **Método de Gauss**:
   - La **suma de Gauss**, por otro lado, tiene una complejidad **constante** \( O(1) \). Esto significa que su tiempo de ejecución debería ser **prácticamente independiente** del tamaño de `n`.
   - Al graficar los tiempos de ejecución, se observa que el tiempo es casi constante, independientemente de los valores de `n`. Incluso para valores muy grandes de `n` (como \( n = 10^6 \) o \( n = 10^7 \)), el tiempo de ejecución no cambia significativamente.
   - En el gráfico de dispersión, los puntos para el algoritmo de Gauss se agrupan en una zona baja de tiempo, lo que refleja su complejidad constante.

### **Análisis Teórico**

El análisis teórico se centra en entender el comportamiento de los algoritmos sin la necesidad de realizar pruebas prácticas. Aquí se evalúa la **complejidad computacional** de cada uno de los métodos:

1. **Método Iterativo**:
   - **Descripción**: En el método iterativo, el algoritmo recorre todos los números de 1 a `n` y los va sumando.
   - **Complejidad**: Este algoritmo tiene una **complejidad temporal lineal** \( O(n) \), ya que el tiempo de ejecución es directamente proporcional al tamaño de la entrada. Esto significa que, si `n` aumenta en un factor de 10, el tiempo de ejecución también aumentará aproximadamente en un factor de 10.
   - **Explicación**: El algoritmo realiza una operación (sumar) para cada número entre 1 y `n`. Si `n = 1000`, se realizarán 1000 sumas, lo que da como resultado un tiempo de ejecución proporcional a `n`.

2. **Método de Gauss**:
   - **Descripción**: El método de Gauss utiliza la fórmula \( S(n) = \frac{n(n + 1)}{2} \) para calcular la suma de los primeros `n` números sin necesidad de realizar ninguna iteración.
   - **Complejidad**: Este algoritmo tiene una **complejidad constante** \( O(1) \), ya que su tiempo de ejecución no depende del tamaño de `n`. La fórmula matemática requiere una cantidad constante de operaciones, independientemente de cuán grande sea `n`.
   - **Explicación**: La fórmula de Gauss no depende de la cantidad de números que estamos sumando. Es una fórmula directa, y por lo tanto, no importa si `n` es 10, 1000 o 10 millones; siempre tomará el mismo tiempo para calcular el resultado.

### **Conclusión**

1. **Comparación de los Métodos**:
   - El **método iterativo** crece proporcionalmente con el tamaño de la entrada, lo que es una característica típica de algoritmos con complejidad **lineal** \( O(n) \). Esto lo hace menos eficiente para valores grandes de `n`, ya que el tiempo de ejecución aumenta considerablemente con el tamaño de la entrada.
   - El **método de Gauss**, por otro lado, tiene una **complejidad constante** \( O(1) \), lo que significa que, independientemente del tamaño de `n`, su tiempo de ejecución será prácticamente el mismo. Esto lo hace muy eficiente, incluso cuando `n` alcanza valores extremadamente grandes.

2. **Análisis Empírico**:
   - Los resultados empíricos confirman la teoría. El gráfico de la suma iterativa muestra un aumento lineal en el tiempo de ejecución con el tamaño de `n`. Mientras tanto, la suma de Gauss mantiene un tiempo de ejecución prácticamente constante a medida que aumenta `n`.
   - La comparación entre ambos métodos destaca la superioridad del algoritmo de Gauss para este problema, ya que es mucho más eficiente.

3. **Recomendaciones**:
   - Para problemas que implican la suma de los primeros `n` números, **usar la fórmula de Gauss es la opción más eficiente**, ya que tiene una complejidad constante y su tiempo de ejecución no aumenta con el tamaño de `n`.
   - El método iterativo, aunque más sencillo de entender y aplicar, es menos eficiente para valores grandes de `n`.

En resumen, el análisis empírico y teórico confirma que, en este caso, la **suma de Gauss** es mucho más eficiente que el **método iterativo**, tanto desde el punto de vista de la teoría de la complejidad como desde la observación empírica del tiempo de ejecución.
